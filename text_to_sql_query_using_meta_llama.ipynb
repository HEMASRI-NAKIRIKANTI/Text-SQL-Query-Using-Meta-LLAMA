````python
# This class provides functionality related to Hugging Face Transformers pipelines .
from langchain import HuggingFacePipeline

# This line imports the AutoTokenizer class from the transformers library.
# The AutoTokenizer class is used to load tokenizers for various pre-trained language models available in the Hugging Face model hub.
from transformers import AutoTokenizer

# This line imports the entire transformers library, which is a popular library developed by
# Hugging Face for working with various transformer-based models in natural language processing (NLP),
# including both models and tokenizers.
import transformers

# This line imports the torch library, which is the primary library used for deep learning and tensor computations in PyTorch.
import torch

# Model name that we want to use
# https://huggingface.co/meta-llama/Llama-2-7b-chat-hf

model = "meta-llama/Llama-2-7b-chat-hf"

access_token ="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"
# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model,token=access_token)
# Set up text generation pipeline
# *************** num_return_sequences************************
# num_return_sequences to 3 and provide an input prompt like "Once upon a time," the model will generate three different continuations or completions of the input prompt.
# Input prompt: "Once upon a time,"
# Generated sequences might look like:
# "Once upon a time, there was a little girl named Alice. She lived in a small cottage in the forest..."
# "Once upon a time, in a faraway kingdom, there lived a brave knight named Sir Arthur..."
# "Once upon a time, the sun was shining brightly, and the birds were chirping merrily in the trees..."
pipeline = transformers.pipeline("text-generation",
                model=model,                             # type of model we are selecting here meta-llama
                tokenizer= tokenizer,                    # It preprocesses raw text data so that it can be properly understood by the model
                torch_dtype=torch.bfloat16,              # model parameters and computations will be performed using the bfloat16 data type, which can be useful for optimizing performance and memory usage,
                device_map="auto",                       # allows the library to choose the appropriate device automatically based on availability.
                max_new_tokens = 512,                    # in addition to the input text provided to the model for text generation, the model can generate up to 512 additional tokens.
                do_sample=True,                          # determines whether to use sampling when generating text. Sampling means selecting the next token based on a probability distribution, which can lead to more diverse outputs
                top_k=10,                                #Input prompt: "I enjoy eating" it has  Vocabulary: ["apple", "banana", "orange", "pear", "grape"].If top_k is set to 2, the model might consider the following options for the next word based on their probabilities:If top_k is set to 2, the model might consider the following options for the next word based on their probabilities:"apple" - probability: 0.35 "banana" - probability: 0.30 In this case, the model would choose between "apple" and "banana" as the next word, as they are the top 2 most probable words in the vocabulary.
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id,
                )
````

````python
# 'HuggingFacePipeline' class creates a custom pipeline for text generation, and we are passing
# the pipeline that we defined earlier along with some model-specific keyword arguments - temperature here.

llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0})
````

````python
from langchain import PromptTemplate,  LLMChain

template = """
             Create a SQL query snippet using the below text:
              ```{text}```
              Just SQL query:
           """

prompt = PromptTemplate(template=template, input_variables=["text"])

llm_chain = LLMChain(prompt=prompt, llm=llm)

text = """ Extract all the unique values from column "age"
"""
````

````python
# use invoke function to see the output response
print(llm_chain.invoke(text))
````
